```python
import os
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
print(os.getcwd())
```

```python
main_data_dir = os.getcwd() + "/drive"
print(os.listdir(main_data_dir))
```

```python
train_dir = main_data_dir + "/training"
test_dir = main_data_dir + "/test"

train_img_dir = train_dir + "/images"
train_mask_dir = train_dir + "/1st_manual"

test_img_dir = test_dir + "/images"

print(f"No. Train Img: {len(os.listdir(train_img_dir))}")
print(f"No. Train Msk: {len(os.listdir(train_mask_dir))}")
print(f"No. Test Img: {len(os.listdir(test_img_dir))}")
```

```python
train_img_paths = glob(train_img_dir + "/*")
train_msk_paths = glob(train_mask_dir + "/*")

idx = 0
sample_img_path, sample_msk_path = train_img_paths[idx], train_msk_paths[idx]

print(sample_img_path)
print(sample_msk_path)
```

```python
sample_img = Image.open(sample_img_path)
sample_msk = Image.open(sample_msk_path)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))

ax1.imshow(np.array(sample_img))
ax2.imshow(np.array(sample_msk), cmap = 'gray')
```
